# Import library

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('putriauliairwan.jpeg')

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 

plt.imshow(img)
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 
```

```python
edged = cv2.Canny(gray, 30, 200) 
cv2.waitKey(0) 
contours, hierarchy = cv2.findContours(edged, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)  
cv2.imshow('Canny Edges After Contouring', edged)
cv2.imwrite('CannyFish.jpg', edged)  
cv2.waitKey(0) 
```

```python
print("Number of Contours: " + str(len(contours))) 
```

```python
cv2.drawContours(img, contours, -1, (0, 255, 0), 3) 
```

```python
cv2.imshow('Contours', img)
cv2.imwrite('Contours.jpg', img) 
cv2.waitKey(0) 
```

```python
img2 = cv2.imread('putriauliairwan.jpg')
if len(contours) != 0:
	c = max(contours, key = cv2.contourArea)
cv2.drawContours(img2, c, -1, (0, 255, 0), 3)
cv2.imshow('Largest Contour', img)
cv2.imwrite('LargestContour.jpg', img) 
cv2.waitKey(0) 

plt.imshow(img)
plt.show()
```
